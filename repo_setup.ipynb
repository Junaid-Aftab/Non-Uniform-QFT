```python
# ============================================================
#   Create a clean, neutral repository layout with zero algorithmic or numerical assumptions.
#   Import Packages
# ============================================================

import os
import sys
from pathlib import Path
```

```python
# ============================================================
#   Ensure this notebook is being executed from the repository root.
#   Many paths below assume the current working directory is the repo root.
# ============================================================

# Show current working directory
cwd = Path.cwd()
print("Current working directory:", cwd)

# OPTIONAL: If you opened Jupyter from a parent folder, you can change into the repo folder.
# Example:
# os.chdir("/Users/junaida/Documents/Non-Uniform-QFT")
# print("New working directory:", Path.cwd())
```

```python
# ============================================================
#   Create the standard repo folders for reproducible experiments:
#     - src/nuqft     : importable Python package code (non-notebook)
#     - notebooks     : Jupyter notebooks (numbered for execution order)
#     - data          : generated datasets
#     - results       : figures/tables generated by experiments
#     - tests         : unit tests for helper code
#     - .github/workflows : CI workflows (optional, but included for reproducibility)
# ============================================================

folders = [
    Path("src/nuqft"),
    Path("src/nuqft/circuits"),
    Path("notebooks"),
    Path("data"),
    Path("results"),
    Path("tests"),
    Path(".github/workflows"),
]

for f in folders:
    f.mkdir(parents=True, exist_ok=True)

print("Created/verified folders:")
for f in folders:
    print(" -", f.resolve())
```

```python
# ============================================================
#   Create minimal Python package initialization so that:
#     - you can `import nuqft` inside notebooks
#     - your code is cleanly separated from notebooks
# ============================================================

init_files = [
    Path("src/nuqft/__init__.py"),
    Path("src/nuqft/circuits/__init__.py"),
]

for fp in init_files:
    if not fp.exists():
        fp.write_text(
            "# nuqft package\n"
            "# Purpose: reusable code for NUQFT/NUDFT experiments.\n"
        )
        print(f"Created {fp}")
    else:
        print(f"Exists  {fp}")
```

```python
# ============================================================
#   Ensure that `src/` is on the Python path when running notebooks.
#   This avoids ad-hoc relative imports and keeps notebooks clean.
#
# Notes:
#   - This pattern is common in research repos.
#   - Later you can switch to `pip install -e .` if preferred.
# ============================================================

src_path = (Path.cwd() / "src").resolve()
if str(src_path) not in sys.path:
    sys.path.insert(0, str(src_path))

print("Added to sys.path:", src_path)
```

```python
# ============================================================
#   Add placeholders so that empty directories can be committed to Git.
# ============================================================

for fp in [Path("data/.gitkeep"), Path("results/.gitkeep")]:
    if not fp.exists():
        fp.write_text("")
        print(f"Created {fp}")
    else:
        print(f"Exists  {fp}")
```
